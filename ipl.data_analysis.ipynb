```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/adars/Downloads/cricket_data.csv")
print("Dataset Preview:")
df.head()
```

```python
print("Dataset Info:")
df.info()
```

```python
print("Statistical Summary:")
df.describe()
```

```python
df["Highest_Score"] = df["Highest_Score"].astype(str).str.replace("*", "")
```

```python
df.head(100)
```

```python
numeric_columns = [
    "Matches_Batted","Not_Outs","Runs_Scored","Highest_Score","Batting_Average",
    "Balls_Faced","Batting_Strike_Rate","Centuries","Half_Centuries",
    "Fours","Sixes","Catches_Taken","Stumpings","Matches_Bowled",
    "Balls_Bowled","Runs_Conceded","Wickets_Taken","Bowling_Average",
    "Economy_Rate","Bowling_Strike_Rate","Four_Wicket_Hauls","Five_Wicket_Hauls"
]
```

```python
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors="coerce")
print("\nData Types After Conversion:")
df.dtypes
```

```python
df.isnull().sum()
```

```python
median_sr = np.median(df["Batting_Strike_Rate"].dropna())
mean_sr = np.mean(df["Batting_Strike_Rate"].dropna())
print(f"Median Batting Strike Rate: {median_sr:.2f}")
print(f"Mean Batting Strike Rate: {mean_sr:.2f}")
```

```python
avg_runs = np.mean(df["Runs_Scored"].dropna())
above_avg_players = df[df["Runs_Scored"] > avg_runs]["Player_Name"].unique()
print(f"\nPlayers who scored above average runs ({avg_runs:.0f}):")
print(above_avg_players)
```

```python
top_scorers = df.groupby("Player_Name")["Runs_Scored"].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Run Scorers:")
print(top_scorers)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(y=top_scorers.index, x=top_scorers.values)
plt.title("Top 10 Run Scorers")
plt.xlabel("Total Runs")
plt.ylabel("Player")
plt.savefig("top_run_scorers.png")
plt.show()
```

```python
top_wickets = df.groupby("Player_Name")["Wickets_Taken"].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Wicket Takers:")
print(top_wickets)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(y=top_wickets.index, x=top_wickets.values)
plt.title("Top 10 Wicket Takers")
plt.xlabel("Total Wickets")
plt.ylabel("Player")
plt.savefig("top_wicket_takers.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Batting_Average"].dropna(), bins=25)
plt.title("Distribution of Batting Averages")
plt.xlabel("Batting Average")
plt.ylabel("Frequency")
plt.savefig("batting_average_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Batting_Strike_Rate", y="Runs_Scored")
plt.title("Strike Rate vs Runs Scored")
plt.xlabel("Batting Strike Rate")
plt.ylabel("Runs Scored")
plt.tight_layout()
plt.savefig("strike_rate_vs_runs.png")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Economy_Rate", y="Wickets_Taken")
plt.title("Economy Rate vs Wickets Taken")
plt.xlabel("Economy Rate")
plt.ylabel("Wickets Taken")
plt.tight_layout()
plt.savefig("economy_vs_wickets.png")
plt.show()
```
